# Guest House Data Purification

```python
import pandas as pd
df=pd.read_csv('Guest_house.csv')
df
```

```python
df=df.set_index('id')
```

```python
df.describe()
```

```python
df.info()
```

# 1. Drop NA Values

```python
df.isnull().sum()
```

```python
df.dropna(subset='name',inplace=True)
df.shape
```

```python
df.dropna(subset='host_name',inplace=True)
df.shape
```

```python
df.dropna(subset='last_review',inplace=True)
df.dropna(subset='reviews_per_month',inplace=True)
df.shape
```

# 2. Sorting

```python
df.sort_index(ascending=False)
```

```python
df.sort_values(['price'],ascending=False)
```
